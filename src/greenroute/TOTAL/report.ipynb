#### **I. Introduction**
The pharmaceutical industry generates large amount of chemical waste during drug synthesis. Green chemistry provides tools to design more sustainable chemical processes.
The key metrics allowing chemmists to determine whether a process is greener than another are atom economy, PMI and E factor. Atom economy measures how efficiently atoms are used in a chemical reaction to minimize waste production, described by the following formula.
$$AE= Number of atoms in the product / Number of atoms in the reactant$$
Process Mass Intensity (PMI) is used to decrease the amount of material used to make a drug in the pharmaceutical industry since it is one of its main challenges. A lower PMI indicates a higher effiency, less waste and therefore, better sustainability, as described by the following equation.
$$PMI= Total mass input in the process / Mass of the product$$
